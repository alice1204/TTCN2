```python
import pandas as pd
import numpy as np

# đường dẫn file
files = {
    "Gold": "gold_prices.csv",
    "Silver": "silver_prices.csv",
    "Oil": "oil_prices.csv",
    "USD": "usd_prices.csv",
    "SP500": "sp500_prices.csv"
}

def read_csv(file_path, date_col_candidates=["Date", "date", "DATE"]):
    df = pd.read_csv(file_path)
    # tìm cột ngày
    date_col = None
    for c in date_col_candidates:
        if c in df.columns:
            date_col = c
            break
    if date_col is None:
        raise ValueError(f"No date column found in {file_path}")
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values(date_col).reset_index(drop=True)
    df = df.rename(columns={date_col: "Date"})
    return df

# đọc tất cả vào dict
data = {name: read_csv(path) for name, path in files.items()}
# kiểm tra nhanh
for name, df in data.items():
    print(name, df.shape)
    display(df.head(2))
```
